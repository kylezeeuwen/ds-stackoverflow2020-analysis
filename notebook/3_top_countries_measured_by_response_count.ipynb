# File Overview

This file lists country by response count and computes the percentage of the responses of the top chosen countries as a percentage of all the responses.

# Outputs / Assets

* response distribuition bar chart for use in the blog
* percentage of responses from top countries

# Data Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 2000
pd.options.display.max_columns = 1000
pd.options.display.max_colwidth = 255

sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sns.set_context('notebook')
sns.set_style("ticks")

df = pd.read_csv('./assets/survey_results_public.csv')
```

## Which countries had the most survey respondents, using a round number of 1000 as the cutoff ?

```python
value_counts = df['Country'].value_counts()
large_countries_by_response = value_counts[value_counts > 1000]
large_countries_by_response
```

```python
(large_countries_by_response).plot(kind="bar")
plt.title("Top Countries measured by Response Count");
```

## What proportion of the total data set do these countries represent ?

```python
round(100 * large_countries_by_response.sum() / df['Country'].value_counts().sum(), 1)
```

# Data Evaluation

The top 13 countries by response volume make up 65.2% of the responses to the 2020 Stack overflow survey.
